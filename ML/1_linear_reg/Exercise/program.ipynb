```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Exercise\canada_per_capita_income.csv")
df.head(3)
```

```python
x = df.drop("per capita income (US$)", axis="columns")
y = df[["per capita income (US$)"]]
model = linear_model.LinearRegression()
model.fit(x,y)
model.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income $')
plt.scatter(df.year,df["per capita income (US$)"],color='red',marker='o')
plt.plot(df.year,model.predict(df[["year"]]), color = "blue")
```
